# Lecture du dataset apres Phase 1

Dataset produit par `01_data_cleaning.ipynb` : `data/processed/dataset_ml_final.parquet` (349 390 lignes x 24 colonnes).

```python
import pandas as pd
from pathlib import Path

pd.set_option('display.max_columns', None)
pd.set_option('display.width', 200)

PATH = Path('..') / 'data' / 'processed' / 'dataset_ml_final.parquet'
df = pd.read_parquet(PATH)
print('Shape :', df.shape)
df.head()
```

## Schema / types

```python
df.dtypes
```

## Valeurs manquantes

```python
na = df.isna().sum()
na[na > 0].sort_values(ascending=False)
```

## Statistiques descriptives (numerique)

```python
df.describe()
```

## Variables categorielles

```python
df.select_dtypes(include=['object', 'category']).describe()
```

## Plage temporelle

```python
date_cols = [c for c in df.columns if 'date' in c.lower()]
for c in date_cols:
    s = pd.to_datetime(df[c], errors='coerce')
    print(f"{c:30s} : {s.min()}  ->  {s.max()}")
```
